```python
import gensim
import pandas as pd
```

```python
import json
import pandas as pd

file_path = "/content/AMAZON_FASHION.json"
data = []

with open(file_path, 'r') as f:
    for line_num, line in enumerate(f):
        try:
            data.append(json.loads(line))
        except json.JSONDecodeError as e:
            print(f"Skipping line {line_num + 1} due to JSONDecodeError: {e}")

df = pd.DataFrame(data)
display(df)
```

```python
df.shape
```

```python
# simple preprocessing and tokenization
df['reviewText'] = df['reviewText'].astype(str)
review_text = df.reviewText.apply(gensim.utils.simple_preprocess)
review_text
```

```python
df['reviewText'].apply(type).value_counts()
```

```python
review_text.loc[2]
```

```python
df.reviewText.loc[2]
```

```python
# initialize the model
model=gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4
)
```

```python
# build vocab
model.build_vocab(review_text, progress_per=1000)
```

```python
# train the model
model.train(review_text, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
# save the model so it can be used in other application
# model.save("./word2vec-/content/AMAZON_FASHION.model") if we have load the model
```

```python
# check similarity between words
model.wv.similarity(w1="cheap", w2="inexpensive")
```
